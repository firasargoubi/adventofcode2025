FUNCTION TO PARSE DATA

```python
def parse(filename) :
    ans = []
    with open(filename,"r") as f :
        for line in f :
            #print(line)
            num = int(line[1:])
            if line[0] == "L" :
                num *= -1
            ans.append(num)

    return ans
```

PART ONE

```python
data = parse("input2.txt")
count = 0
start = 50 
for num in data :
    start = (start + num) % 100
    if not start :
        count+=1
print(count)
        
```

PART TWO

```python
data = parse("input.txt")
count = 0
start = 50 
for num in data :
    n = (start + num)
    turns = abs(n) // 100
    count+= turns
    if n <= 0 and start != 0 :
        count += 1

    start = n % 100
print(count)
        
```
