FUNCTION TO PARSE DATA

```python
def parse(filename) :
    s = []
    l = []
    c = False
    with open(filename,"r") as f :
        
        for line in f :
            #print(line)
            if line == "\n" :
                c = True
                continue
            if c :
                l.append(int(line.strip()))
            else :
                right,left = map(int,line.strip().split("-"))
                s.append([right,left])
    s.sort(key = lambda x : x[0])
    return s,l
```

PART ONE

```python
valid,ids = parse("input.txt")
count = 0
for j in ids :
    for r,l in valid :
        if r <= j <= l :
            count+=1
            break

print("Count : ",count)
```

PART TWO

```python
valid,ids = parse("input.txt")
new = []
l,r = valid[0]
i =1 
while i < len(valid) :
    nL,nR = valid[i]
    if nL <= r  :
        r = max(r,nR)
        if i == len(valid) -1 :
            new.append([l,r])
    else :
        new.append([l,r])
        r,l = nR,nL
        if i == len(valid) -1 :
            new.append([l,r])
    i+=1
count = 0
for a,b in new :
    count += b - a + 1
print("Count : ",count)
```
