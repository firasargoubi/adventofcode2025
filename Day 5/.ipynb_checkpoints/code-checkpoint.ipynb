FUNCTION TO PARSE DATA

```python
def parse(filename) :
    s = []
    l = []
    c = False
    with open(filename,"r") as f :
        
        for line in f :
            #print(line)
            if line == "\n" :
                c = True
                continue
            if c :
                l.append(int(line.strip()))
            else :
                right,left = map(int,line.strip().split("-"))
                s.append([right,left])
    s.sort(key = lambda x : x[0])
    return s,l
```

PART ONE

```python
valid,ids = parse("input.txt")
count = 0
for j in ids :
    for r,l in valid :
        if r <= j <= l :
            count+=1
            break

print("Count : ",count)
```

PART TWO

```python
valid, _ = parse("test.txt")
new = []
i = 1
r,l = valid[0]
while i < len(valid) :
    nR,nL = valid[i]
    if r <= nR <= l and l < nL:
        l = nL
    else :
        new.append([r,l])
        r,l = nR,nL
    print(i,new)
    i +=1

        
```
