FUNCTION TO PARSE DATA

```python
def parse(filename) :
    ans = []
    with open(filename,"r") as f :
        lines = f.readlines()
        operations = lines[-1]
        indexes = []
        for i in range(len(operations)) :
            if operations[i] in ("*","/","+","-") :
                indexes.append(i)
        for line in lines :
            t = []
            for j in range(len(indexes)-1) :
                l,r = indexes[j:j+2]
                t.append(line[l:r-1])
            t.append(line[indexes[-1]:-1])
            ans.append(t)

    return ans
```

PART ONE

```python
data = parse("input.txt")
operations = [a.strip() for a in data[-1]]
counts = [int(b) for b in data[0]]
for a in data[1:-1] :
    for i in range(len(a)) :
        if operations[i] == "*" :
            counts[i] *= int(a[i])

        elif operations[i] == "/" :
            counts[i] = counts[i] / int(a[i])

        elif operations[i] == "-" :
            counts[i] -= int(a[i])

        elif operations[i] == "+" :
            counts[i] += int(a[i])
count = sum(counts)
print("Count : ", count)
    
```

PART TWO

```python
data = parse("input.txt")
#invert the data
inverted = []
for i in range(len(data[0])) :
    t = []
    for j in range(len(data)-1) :
        t.append(data[j][i])
    inverted.append(t)

#calculate the new values right to left
new = [[0]* len(a[0]) for a in inverted]
for k in range(len(inverted)) :
    for elem in inverted[k] :
        for i in range(len(elem)) :
            if elem[-(i+1)] == " ":
                continue
            new[k][i] = new[k][i] * 10 + int(elem[-(i+1)])

#same as before
operations = [a.strip() for a in data[-1]]
counts = [int(b[0]) for b in new]
for k in range(len(new)) :
    if operations[k] == "*" :
        for elem in new[k][1:] :
            counts[k] *= elem

    elif operations[k] == "/" :
        for elem in new[k][1:] :
            counts[k] /= elem

    elif operations[k] == "-" :
        for elem in new[k][1:] :
            counts[k] -= elem

    elif operations[k] == "+" :
        for elem in new[k][1:] :
            counts[k] += elem

count = sum(counts)
print("Count : ", count)    
```
