FUNCTION TO PARSE DATA

```python
def parse(filename) :
    ans = []
    with open(filename,"r") as f :
        for line in f :
            ans.append(line.strip())

    return ans
```

PART ONE

```python
data = parse("input.txt")
start = data[0].index("S")
count = 0
queue = [start]
for line in data[1:] :
    n = set()
    for elem in queue :
        if line[elem] != "." :
            count +=1 
            n.add(elem-1)
            n.add(elem+1)
        else :
            n.add(elem)
    queue = [a for a in n]
print("Count : ",count)
    
```

PART TWO

```python
data = parse("input.txt")
start = data[0].index("S")
maxX = len(data)
maxY = len(data[0])
count = 0
r = [[0 for _ in range(maxY)] for _ in range(maxX)]
def recur(level,index) :
    if level == maxX-1 :
        #print("Success",level,index)
        r[level][index] = 1
        return 1
    if r[level][index] > 0:
        return r[level][index]
    v = 0
    if data[level][index] == "." :
        cur = recur(level+1,index)
        v+=cur
    else :
        left = recur(level,index-1)
        right = recur(level,index+1)
        v+= left + right
    r[level][index] = v
    return v
val = recur(1,start)
print("Count : ",val)
    
```
