FUNCTION TO PARSE DATA

```python
def parse(filename) :
    ans = []
    with open(filename,"r") as f :
        for line in f:
            coords = list(map(int,line.strip().split(",")))
            ans.append(coords)

    return ans
```

```python
data = parse("input.txt")
connections = []
links = [set([i]) for i in range(len(data))]
for i in range(len(data)) :
    x,y,z = data[i]
    for j in range(i+1,len(data)) :
        aX,aY,aZ = data[j]
        dif = ((x - aX)**2 + (y - aY)**2 + (z - aZ )**2) ** 0.5
        connections.append((f"{i}-{j}",dif))
connections.sort(key = lambda x : x[1])
for t in range(1000) :
    key,_ = connections[t]
    left,right = list(map(int,key.split("-")))
    l , r = -1,-1
    for k,elem in enumerate(links) :
        if left in elem :
            l = k
        if right in elem : 
            r = k
    if l != r :
        if r < l :
            temp = r 
            r = l
            l = temp
        
        add = links.pop(r)
        links[l] = links[l].union(add)
sizes = [len(s) for s in links]
sizes.sort(reverse = True)
count = 1
for m in range(3) :
    count *= sizes[m]
print("Count : ",count)
```
