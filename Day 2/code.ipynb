FUNCTION TO PARSE DATA

```python
def parse(filename) :
    ans = []
    with open(filename,"r") as f :
        ans = f.read().split(",")
    return ans
```

PART ONE

```python
data = parse("input.txt")
print(data)
count = 0
for interval in data :
    left,right = interval.split("-")
    for i in range(int(left),int(right)) :
        t = f"{i}"
        if (len(t) % 2 != 0) :
            continue
        length = len(t) // 2
        if (t[:length] == t[length:]) :
            count += i
print("COUNT : ",count)
```

PART TWO

```python
data = parse("input.txt")
def combinations(string,mult) :
    if (len(string) % mult != 0 ) :
        return []
    
    c = len(string) // mult
    ans = []
    for i in range(mult) :
        ans.append(string[c*i:c*(i+1)])
    return ans

def checkArray(arr) :
    c = "-".join(arr[0])
    for elem in arr :
        if c != "-".join(elem) :
            return False
    return True
print(data)
count = 0
for interval in data :
    left,right = interval.split("-")
    for i in range(int(left),int(right)+1) :
        t = f"{i}"
        length = len(t)
        for j in range(2,length+1) :
            arr = combinations(t,j)
            if not len(arr) :
                continue
            if checkArray(arr) :
                count += i
                break
print("COUNT : ",count)
```
