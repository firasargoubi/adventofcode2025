FUNCTION TO PARSE DATA

```python
def parse(filename) :
    ans = []
    with open(filename,"r") as f :
        for line in f :
            temp = []
            for char in line :
                if char == "\n" :
                    continue
                temp.append(int(char))
            ans.append(temp)

    return ans
```

PART ONE

```python
data = parse("input.txt")
count = 0
for array in data :
    m1 = 0
    m2 = 0
    num = 0
    for i in range(len(array)-1,-1,-1) :
        if m2 == 0 :
            m2 = array[i]
        elif m1 == 0 :
            m1 = array[i]
            num = m1 * 10 + m2
        else :
            if m2 < array[i+1] :
                m2 = array[i+1]
                m1 = array[i]
            if m1 < array[i] :
                m1 = array[i]
            t = m1*10 + m2
            if t > num :
                num = t
    count +=num
print("Count : ",count)
```

```python
data = parse("input.txt")
count = 0
dp = {}

def recur(n, left, array):
    state = (n, left)
    if state in dp:
        return dp[state]
    
    rem = len(array) - left
    if rem < n:
        return -1
    if n == 1:
        return max(array[left:])
    if n == rem:
        val = 0
        for i in range(left, len(array)):
            val = val * 10 + array[i]
        return val

    pick = recur(n-1, left+1, array)
    cur = -1
    if pick != -1:
        cur = (array[left] * (10**(n-1))) + pick
    skip = recur(n, left+1, array)

    best = max(cur, skip)
    dp[state] = best
    return best

for row in data:
    dp = {}
    m = recur(12, 0, row)
    count += m

print("COUNT : ",count)
    
```

a revoir avec un stack a la place ,a chaque fois on retire du stack tant que cur index > top of stack et max on retire len - 12 elems

```python
data = parse("input.txt")
count = 0

for array in data :
    toDelete = len(array) - 12
    
    stack = []
    for num in array :
        while stack and stack[-1] < num and toDelete > 0 :
            stack.pop()
            toDelete -= 1
        stack.append(num)
        
    count += int(''.join([str(k) for k in stack[:12]]))

print("COUNT : ",count)
```
